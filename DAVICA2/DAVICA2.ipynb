```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import re
from matplotlib.lines import Line2D
    
carPopulation = pd.read_csv('carPopulation.csv',sep= ',')
#carPopulation['Year'].drop_values()
carPopulation = carPopulation.drop(carPopulation[carPopulation.Year < 2000].index)
carPopulation.set_index('Year')
sns.set(rc = {'figure.figsize':(16,9)})
#carpopline = sns.lineplot(data=carPopulation, x="Year", y = "Private & Company Cars" )
averageCarMileage = pd.read_csv('averageCarMileage.csv',sep= ',',index_col=0)
g = sns.lineplot(data=carPopulation,x="Year", y = "Private & Company Cars", color="g")
sns.lineplot(data=averageCarMileage,x="Year", y="Average Annual Mileage", color="b", ax=g.axes.twinx())
g.legend(handles=[Line2D([], [], marker='_', color="g", label='Number of Private & Company Cars'), Line2D([], [], marker='_', color="b", label='Average Annual Mileage of cars')])
#carmileline = sns.lineplot(data=averageCarMileage, x="Year", y="Average Annual Mileage")
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import re
import scipy as sp
averageMRTRidership = pd.read_csv('averageMRTRidership.csv',sep= ',',index_col=0)
averageMRTDistance = pd.read_csv('averageMRTDistance.csv',sep= ',',index_col=0)
noofMRT = pd.read_csv('noofMRT.csv',sep= ',')
#e = sns.histplot(data = noofMRT, x = 'Year', y = 'number of MRT Stations', color = 'r') 
#sns.barplot(data = noofMRT, x = 'Year', y = 'number of MRT Stations')
g = sns.lineplot(data = averageMRTRidership, x = 'Year', y = 'Average Daily Ridership - MRT (Thousand Passenger-trips)', color="g")
sns.lineplot(data=averageMRTDistance,x="Year", y="Average Distance per Trip", color="b", ax=g.axes.twinx())
g.legend(handles=[Line2D([], [], marker='_', color="g", label='Average Daily Ridership - MRT (Thousand Passenger-trips)'), Line2D([], [], marker='_', color="b", label='Average Distance per Trip')])
```
